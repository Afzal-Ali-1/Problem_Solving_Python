```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from matplotlib.animation import FuncAnimation
import psutil
import time
```

```python
%matplotlib notebook
fig=plt.figure()
ax=fig.add_subplot(111)
y=[psutil.cpu_percent() for var in range(50)]
x=[var for var in range(50)]
ax.plot(x,y)
def change(i):
    global x,y 
    x.pop(0)
    y.pop(0)
    x.append(x[-1]+1)
    y.append(psutil.cpu_percent())
    ax.clear()
    ax.plot(x,y,'ro--')
ani=FuncAnimation(fig,change,2000)
plt.show()
```

```python
x = np.arange(1,11)
y = np.random.randint(20,50,10)
print(x)
print(y)
plt.scatter(x,y)

print(x.mean())
print(y.min())
print(y.max())
print(x.std())
print([x.mean()-x.std()]*2)
plt.plot([x.mean(),x.mean()],[y.min(),y.max()],c='r',lw=2)
plt.plot([x.mean()-x.std()]*2,[y.min(),y.max()],c='g',lw=2)
plt.plot([x.mean()+x.std()]*2,[y.min(),y.max()],c='g',lw=2)
for index in range(len(x)):
    plt.plot([x[index],x.mean()],[y[index]]*2,'r--',lw=2)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np 
from time import sleep
from psutil import cpu_percent as cpu
%matplotlib notebook
x = [var for var in range(1,50)]
y = [var for var in range(1,50)]
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(x,y,'go--')
ax.set_xlim([1,15])
ax.set_ylim([1,15])
plt.show()
while True : 
    x.pop(0)
    y.pop(0)
    x.append(x[-1]+1)
    y.append(cpu())
    ax.clear()
    ax.plot(x,y,'go--')
    fig.canvas.draw()
    sleep(1)
```
